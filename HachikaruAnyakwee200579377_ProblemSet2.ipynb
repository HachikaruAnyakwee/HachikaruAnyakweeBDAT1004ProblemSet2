```python
# Define a function 'calculate_file_length' that takes a 'filename' as input
def calculate_file_length(filename):
    try:
        # Attempt to open the file
        with open(filename, 'r') as infile:
            # Read the contents of the file
            file_contents = infile.read()
            # Calculate and print the length of the file contents
            print(f"Length of the file '{filename}': {len(file_contents)} characters")
    except FileNotFoundError:
        # Handle the case where the file is not found
        print(f"File '{filename}' not found.")
    except Exception as e:
        # Handle other exceptions gracefully
        print(f"An error occurred while reading the file: {str(e)}")

# Call the function to calculate the length of 'SampleFile.txt'
calculate_file_length("SampleFile.txt")

# Call the function to calculate the length of 'SampleFile2.txt'
calculate_file_length("SampleFile2.txt")

# Output Explanation
print("\n--- Output Explanation ---")
print("The function 'fileLength' takes a filename as input and prints the length of the file's contents.")

print("\nWhen 'fileLength' is called with 'SampleFile.txt', I attempt to open the file and read its contents.")
print("The 'with' statement is used to automatically close the file after reading.")
print("The length of the file's contents is then printed.")

print("\nIf the file 'SampleFile.txt' is not found, a FileNotFoundError is raised, and I display an appropriate error message.")

print("\nSimilarly, when 'fileLength' is called with 'SampleFile2.txt', I follow the same process.")

print("\nIf any other error occurs while reading the file, I catch it using the 'except' block and display an error message.")
```

```python
Question 3
Write a class named Marsupial that can be used as shown below:
>>> m = Marsupial()
>>> m.put_in_pouch('doll')
>>> m.put_in_pouch('firetruck')
>>> m.put_in_pouch('kitten')
>>> m.pouch_contents() ['doll', 'firetruck', 'kitten']
Now write a class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the y- axis, and
c. overloads the __str__ operator so it behaves as shown below.
>>> k = Kangaroo(0,0)
>>> print(k)
I am a Kangaroo located at coordinates (0,0)
>>> k.put_in_pouch('doll')
>>> k.put_in_pouch('firetruck')
>>> k.put_in_pouch('kitten')
>>> k.pouch_contents() ['doll', 'firetruck', 'kitten']
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> print(k)
I am a Kangaroo located at coordinates (3,0)
```

```python
# Define a class called Marsupial
class Marsupial:
    def __init__(self):
        # Initialize an empty list to represent the pouch of the marsupial
        self.pouch = []
        
    # Method to put an item into the pouch
    def put_in_pouch(self, item):
        self.pouch.append(item)
        
    # Method to get the contents of the pouch
    def pouch_contents(self):
        return self.pouch

# Create an instance of the Marsupial class
m = Marsupial()
# Put items into the pouch
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
# Print the contents of the pouch
print(m.pouch_contents())

# Define a class called Kangaroo that inherits from Marsupial
class Kangaroo(Marsupial):
    def __init__(self, x=0, y=0):
        # Call the constructor of the parent class (Marsupial) using super()
        super().__init__()
        # Initialize the Kangaroo's coordinates
        self.x = x
        self.y = y
        
    # Method to make the Kangaroo jump to new coordinates
    def jump(self, dx, dy):
        self.x += dx
        self.y += dy
        
    # Method to provide a string representation of the Kangaroo
    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"

# Create an instance of the Kangaroo class with initial coordinates (0,0)
k = Kangaroo(0,0)
# Print the Kangaroo's coordinates using the __str__ method
print(k)  

# Put items into the Kangaroo's pouch
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
# Print the contents of the Kangaroo's pouch
print(k.pouch_contents())  

# Make the Kangaroo jump to new coordinates
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
# Print the Kangaroo's updated coordinates
print(k)
```

```python
Question 4
Write function collatz() that takes a positive integer x as input and prints the Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying this rule to the previous number x in the sequence:
x = { 𝑥/2 𝑖𝑓 𝑥 𝑖𝑠 𝑒𝑣𝑒𝑛 3𝑥 + 1 𝑖𝑓 𝑥 𝑖𝑠 𝑜𝑑𝑑
Your function should stop when the sequence gets to number 1. Your implementation must be recursive, without any loops.
>>> collatz(1) 1
>>> collatz(10) 10
5
16
8
4
2
1
```

```python
# Define a function 'collatz' to perform the Collatz sequence
def collatz(x):
    print(x)
    
    # Base case: If 'x' reaches 1, stop the recursion
    if x == 1:
        return
    
    # Recursive case: If 'x' is even, divide it by 2
    elif x % 2 == 0:
        collatz(x // 2)
    
    # Recursive case: If 'x' is odd, apply the Collatz formula
    else:
        collatz(3 * x + 1)

# Call the 'collatz' function with different inputs
collatz(1)
print("---------")
collatz(10)

# Output Explanation
print("\n--- Output Explanation ---")
print("The 'collatz' function performs the Collatz sequence for a given input 'x'.")

print("\nWhen 'collatz' is called with '1', it prints '1' and returns, as it is the base case. The sequence ends here.")

print("\nWhen 'collatz' is called with '10', it prints a sequence of numbers:")
print("10, 5, 16, 8, 4, 2, 1.")
print("The sequence continues until '1' is reached, and each number is printed in the process.")
```

```python
Question 5
Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.
>>> binary(0) 0
>>> binary(1) 1
>>> binary(3) 11
>>> binary(9) 1001
```

```python
# Define a function 'binary' to convert a decimal number to binary
def binary(n):
    # Check if 'n' is greater than 1
    if n > 1:
        # Recursively call 'binary' to divide 'n' by 2
        binary(n // 2)
    
    # Print the remainder of 'n' when divided by 2, without a newline
    print(n % 2, end='')

# Call the 'binary' function with different inputs
binary(0)
print("\n")
binary(1)
print("\n")
binary(3)
print("\n")
binary(9)

# Output Explanation
print("\n--- Output Explanation ---")
print("The 'binary' function converts a decimal number to its binary representation.")

print("\nWhen 'binary' is called with '0', it directly prints '0' as the binary representation of '0'.")

print("\nWhen 'binary' is called with '1', it directly prints '1' as the binary representation of '1'.")

print("\nWhen 'binary' is called with '3', it performs the following steps:")
print("1. 'binary(3 // 2)' is called, which is 'binary(1)'.")
print("2. 'binary(1 // 2)' is called, which is 'binary(0)'.")
print("3. 'binary(0)' is called, and it directly prints '0'.")
print("4. 'binary(1)' resumes and prints '1'.")
print("5. Finally, the binary representation of '3' is '11'.")

print("\nWhen 'binary' is called with '9', it performs the following steps:")
print("1. 'binary(9 // 2)' is called, which is 'binary(4)'.")
print("2. 'binary(4 // 2)' is called, which is 'binary(2)'.")
print("3. 'binary(2 // 2)' is called, which is 'binary(1)'.")
print("4. 'binary(1 // 2)' is called, which is 'binary(0)'.")
print("5. 'binary(0)' is called, and it directly prints '0'.")
print("6. 'binary(1)' resumes and prints '1'.")
print("7. 'binary(2)' resumes and prints '0'.")
print("8. Finally, the binary representation of '9' is '1001'.")
```

```python
Question 6
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. 
You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. 
When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. 
Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.
>>> infile = open('w3c.html')
>>> content = infile.read()
>>> infile.close()
>>> hp = HeadingParser()
>>> hp.feed(content) W3C Mission Principles
```

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.indentation = 0  # Initialize indentation level
        self.headings = []  # List to store parsed headings

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h1') or tag.startswith('h2'):
            # Calculate the indentation level based on the tag (e.g., 'h1' becomes indentation 0)
            self.indentation = int(tag[1:]) - 1

    def handle_data(self, data):
        if self.indentation >= len(self.headings):
            # If the current indentation level is beyond the headings list length, add a new heading
            self.headings.append(data.strip())
        else:
            # If the heading already exists at this indentation level, concatenate the data
            self.headings[self.indentation] += " " + data.strip()

    def print_headings(self):
        for heading in self.headings:
            print(heading + '\n\n')  # Print each heading with line breaks

# Read the content from the specified file
file_path = '/Users/hachikaruanyakwee/Library/CloudStorage/OneDrive-Personal/Documents/GEORGIAN COURSEWORK/BDAT 1004/Problem Set 2/w3c.txt'
with open(file_path, 'r') as infile:
    content = infile.read()

# Create a HeadingParser instance and feed it the content
hp = HeadingParser()
hp.feed(content)

# Print the parsed headings
hp.print_headings()

# Output Explanation
print("\n--- Output Explanation ---")
print("The code uses the 'HeadingParser' class to parse and extract headings from an HTML document.")

print("\n1. The 'handle_starttag' method is used to determine the indentation level based on the HTML tag. For example, 'h1' tags are considered as indentation level 0.")

print("\n2. The 'handle_data' method processes the text data within the tags. If the current indentation level is beyond the headings list length, it adds a new heading. Otherwise, it concatenates the text data with the existing heading.")

print("\n3. The 'print_headings' method prints the parsed headings, with each heading separated by two newline characters.")

print("\n4. The code reads the content of the specified HTML file, creates an instance of 'HeadingParser,' feeds it the content, and finally prints the parsed headings with appropriate formatting.")
```

```python
Question 7
Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.
>>>
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
, 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html http://reed.cs.depaul.edu/lperkovic/csc242/test2.html
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
```

```python
import requests
from bs4 import BeautifulSoup

def webdir(url, depth, indent):
    # Base case: If depth is 0, return without visiting the URL
    if depth == 0:
        return

    try:
        # Send a GET request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for HTTP errors
        soup = BeautifulSoup(response.text, 'html.parser')

        # Print the URL with the appropriate indentation
        print(" " * indent + url)

        # Find all the links on the current web page
        links = soup.find_all('a', href=True)

        # Recursively visit the linked pages with reduced depth and increased indentation
        for link in links:
            new_url = link['href']
            if not new_url.startswith('http'):
                # Handle relative URLs by joining them with the base URL
                new_url = '/'.join(url.split('/')[:-1]) + '/' + new_url
            webdir(new_url, depth - 1, indent + 4)

    except requests.exceptions.RequestException as e:
        # Handle request exceptions
        print("Error:", e)
    except Exception as e:
        # Handle other exceptions
        print("An error occurred:", e)
webdir('http://web.simmons.edu/~grovesd/comm244/notes/week3/html-test-page.html', 2, 0)
```

```python
Question 8
Write SQL queries on the below database table that return:
a) All the temperature data.
b) All the cities, but without repetition.
c) All the records for India.
d) All the Fall records.
e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.
f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.
g) The total annual rainfall for Cairo.
h) The total rainfall for each season.
```

```python
import sqlite3

# Create a SQLite database and connect to it
conn = sqlite3.connect('Weatherdata.db')

# Create a table called WeatherTable
conn.execute('''CREATE TABLE WeatherTable
            (City TEXT, Country TEXT, Season TEXT, Temperature FLOAT, Rainfall FLOAT)''')

# Insert data into the WeatherTable
conn.execute("INSERT INTO WeatherTable (City, Country, Season, Temperature, Rainfall) VALUES ('Mumbai', 'India', 'Winter', 24.8, 5.9)")
conn.execute("INSERT INTO WeatherTable (City, Country, Season, Temperature, Rainfall) VALUES ('Mumbai', 'India', 'Spring', 28.4, 16.2)")
# Continue adding data for Mumbai, London, and Cairo

# Query 1: All the temperature data
query1 = conn.execute("SELECT * FROM WeatherTable")
print("Query 1: All the temperature data.")
for row in query1:
    print(f"City: {row[0]}, Temperature: {row[3]}")

# Query 2: All the cities without repetition
query2 = conn.execute("SELECT DISTINCT City FROM WeatherTable")
print("\nQuery 2: All the cities, but without repetition.")
for row in query2:
    print(f"City: {row[0]}")

# Query 3: All the records for India
query3 = conn.execute("SELECT * FROM WeatherTable WHERE Country='India'")
print("\nQuery 3: All the records for India.")
for row in query3:
    print(f"{row}")

# Query 4: All the Fall records
query4 = conn.execute("SELECT * FROM WeatherTable WHERE Season='Fall'")
print("\nQuery 4: All the Fall records.")
for row in query4:
    print(f"{row}")

# Query 5: The city, country, and season with average rainfall between 200 and 400 millimeters
query5 = conn.execute("""SELECT City, Country, Season
            FROM WeatherTable
            GROUP BY City, Country, Season
            HAVING AVG(Rainfall) BETWEEN 200 AND 400""")
print("\nQuery 5: The city, country, and season for which the average rainfall is between 200 and 400 millimeters.")
for row in query5:
    print(f"{row}")

# Query 6: The city and country with average Fall temperature above 20 degrees in ascending temperature order
query6 = conn.execute("""SELECT City, Country
            FROM WeatherTable
            WHERE Season='Fall'
            GROUP BY City, Country
            HAVING AVG(Temperature) > 20
            ORDER BY AVG(Temperature) ASC""")
print("\nQuery 6: The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order.")
for row in query6:
    print(f"{row}")

# Query 7: Total annual rainfall for Cairo
query7 = conn.execute("SELECT SUM(Rainfall) AS Total_Annual_Rainfall FROM WeatherTable WHERE City='Cairo'")
print("\nQuery 7: The total annual rainfall for Cairo.")
for row in query7:
    print(f"{row}")

# Query 8: Total rainfall for each season
query8 = conn.execute("SELECT Season, SUM(Rainfall) AS Total_Rainfall FROM WeatherTable GROUP BY Season")
print("\nQuery 8: The total rainfall for each season.")
for row in query8:
    print(f"{row}")

# Commit changes and close the connection
conn.commit()
conn.close()
```
